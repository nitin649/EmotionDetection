```python
import numpy as np
import os
import cv2
from skimage.feature import hog
import random
import matplotlib.pyplot as plt
```

```python
lis=os.listdir("E:/Python/KaggleFaceEmotion/face-expression-recognition-dataset/images/train")
```

```python
category=['angry','happy','sad','surprise','neutral']
training_data=[]
```

```python
for i in category:
    image_lis=os.listdir("E:/Python/KaggleFaceEmotion/face-expression-recognition-dataset/images/train/{}".format(i))
    random.shuffle(image_lis)
    final_data=image_lis[0:8000]
    print(len(final_data))
    print(i)
    for j in final_data:
        #print(j)
        data=cv2.imread("E:/Python/KaggleFaceEmotion/face-expression-recognition-dataset/images/train/{}/{}".format(i,j),cv2.IMREAD_GRAYSCALE)
        #feat=hog(data,orientations=9,pixels_per_cell=(8,8),cells_per_block=(1,1),visualize=False,multichannel=False)
        #using hog technique we can extract only import features
        #print(feat)
        final_data=data.reshape(1,-1)
        #print(final_data)
        #final_data=cv2.resize(data,(50,50))
        training_data.append([final_data,category.index(i)])
```
